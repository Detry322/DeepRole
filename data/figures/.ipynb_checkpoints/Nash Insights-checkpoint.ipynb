```python
%matplotlib inline

import numpy as np
import pandas as pd
import gzip
import glob
import matplotlib.pyplot as plt
from collections import defaultdict
import itertools
import seaborn as sns
import swifter
import functools
from scipy.special import comb

BOTS = [
    'RandomBot', 'ObserveBot', 'CFRBot_3000000', 'CFRBot_6000000', 'ISMCTSBot', 'MOISMCTSBot', 'Deeprole',
    'Deeprole_3_1', 'Deeprole_10_5', 'Deeprole_30_15', 'Deeprole_100_50',
    'Deeprole_100_0', 'Deeprole_100_25', 'Deeprole_100_75'
]
# BOTS = ['Deeprole_3_1', 'Deeprole_10_5', 'Deeprole_30_15', 'Deeprole_100_50']
NUM_PLAYERS = 5
```

```python
def load_dataframes():
    result = []
    for filename in glob.glob('tournaments/*.msg.gz'):
        with gzip.open(filename, 'r') as f:
            result.append(pd.read_msgpack(f))
    df = pd.concat(result)
    df.win_type = df.win_type.astype('category')
    df.winner = df.winner.astype('category')
    for i in range(5):
        df['bot_{}'.format(i)] = df['bot_{}'.format(i)].astype('category')
        df['bot_{}_role'.format(i)] = df['bot_{}_role'.format(i)].astype('category')
    df.reset_index(drop=True, inplace=True)
    return df

ALL_GAMES = load_dataframes()
```

```python
def group_by_bot(unchunked_games, bots=BOTS, chunksize=100000):
    results = []
    for i in range(0, len(unchunked_games), chunksize):
        print i, len(unchunked_games)
        games = unchunked_games[i:i+chunksize]
        new_dataframe = {}
        for bot in bots:
            for i in range(5):
                selector = games['bot_{}'.format(i)] == bot
                if i == 0:
                    bot_count = selector.astype(int)
                    payoff = games['bot_{}_payoff'.format(i)][selector]
                else:
                    bot_count = bot_count.add(selector.astype(int), fill_value=0.0)
                    payoff = payoff.add(games['bot_{}_payoff'.format(i)][selector], fill_value=0.0)
            new_dataframe['{}_count'.format(bot)] = bot_count
            new_dataframe['{}_payoff'.format(bot)] = payoff.divide(bot_count)
        results.append(pd.DataFrame(new_dataframe))
    return pd.concat(results)

# GROUPED_GAMES = group_by_bot(ALL_GAMES)
```

```python
def create_payoff_table(grouped_games, bots):
    payoff = grouped_games.groupby(['{}_count'.format(bot) for bot in bots]).mean()
    return payoff[['{}_payoff'.format(bot) for bot in bots]].fillna(0.0)

PAYOFF_TABLE = create_payoff_table(GROUPED_GAMES, BOTS)
import functools

def filter_payoff_table(big_table, bots):
    index_reset = big_table.reset_index()
    interesting_columns = [ col for col in index_reset.columns if any([bot in col for bot in bots]) ]
    filtered = index_reset[interesting_columns]
    criteria = [filtered['{}_count'.format(bot)] != 0 for bot in bots]
    new_criter = functools.reduce(lambda a,b: a | b, criteria)
    bot_sum = functools.reduce(lambda a,b: a + b, [filtered['{}_count'.format(bot)] for bot in bots])
    
    return filtered[new_criter & (bot_sum == NUM_PLAYERS)].groupby(['{}_count'.format(bot) for bot in bots]).mean()
    

table = filter_payoff_table(PAYOFF_TABLE, BOTS)
table
# DEEPROLE_PAYOFFS = {
#     bot: get_payoff_table(PAYOFF_TABLE, ['Deeprole', bot])
#     for bot in BOTS if bot != 'Deeprole'
# }

# DEEPROLE_PAYOFFS['CFRBot_3000000']
```

```python
def scipy_multinomial(params):
    if len(params) == 1:
        return 1
    coeff = (comb(np.sum(params), params[-1], exact=True) *
             scipy_multinomial(params[:-1]))
    return coeff


def P(N_i, x):
    x = np.array(x)
    N_i = np.array(N_i)
    return scipy_multinomial(N_i) * np.prod( x ** N_i )


def r(x, table):
    x = np.array(x)
    numerator = np.zeros(len(x))
    for index, payoff in table.iterrows():
        numerator += P(index, x) * np.array(payoff)
    denominator = 1.0 - (1.0 - x) ** NUM_PLAYERS
    return np.nan_to_num(numerator / denominator)


def calc_xdot(x, table):
    rx = r(x, table)
    xtAx = np.sum(x * rx)
    xdot = x * (rx - xtAx)
    return xdot


LEARNING_RATE = 0.1
def find_nash(table, num_bots=4, start=None, iters=100000, freq=300):
    if start is not None:
        x = start[:]
    else:
        x = np.ones(num_bots) / num_bots
    for i in range(iters):
        x += LEARNING_RATE*calc_xdot(x, table)
        if i % freq == 0:
            print x
```

```python
def two_to_three(vec):
    return np.matmul([[0, 2.0/(3 ** 0.5)], [1, -1/(3 ** 0.5)], [-1, -1/(3 ** 0.5)]], vec) + np.array([0, 0.5, 0.5])

def three_to_two(vec):
    return np.matmul(np.array([[0.5, 1, 0], [1.5 * (3 ** 0.5), (3 ** 0.5), (3 ** 0.5)]]), vec - np.array([0, 0.5, 0.5]))
```

```python
def calc_arrow(x, y, table):
    if y < 0:
        return 0.0, 0.0
    if y > (3 ** 0.5) * (x + 0.5):
        return 0.0, 0.0
    if y > - (3 ** 0.5) * (x - 0.5):
        return 0.0, 0.0
    
    probs = two_to_three(np.array([x, y]))
    
    new_probs = probs + 0.01 * calc_xdot(probs, table)
    new_xy = three_to_two(new_probs)
    return new_xy[0] - x, new_xy[1] - y
```

```python
def make_triangle_plot(big_table, bots):
    table = get_payoff_table(big_table, bots)
    X = np.arange(-0.51, 0.51, 0.02)
    Y = np.arange(0, (3 ** 0.5) / 2, 0.02)
    U, V = np.meshgrid(X, Y)
    
    for i in range(len(U)):
        for j in range(len(U[0])):
            U[i][j], V[i][j] = calc_arrow(U[i][j], V[i][j], table)
    
    color = np.sqrt(U * U + V * V)
    
    plt.figure(figsize=(16,13))
    plt.quiver(X, Y, U/color, V/color, color, scale=60)
    plt.text(-0.5, -0.02, bots[2], horizontalalignment='center')
    plt.text(0.5, -0.02, bots[1], horizontalalignment='center')
    plt.text(0, 0.885, bots[0], horizontalalignment='center')
```

```python
import itertools
def splat_generate_games(games, bots):
    for i in range(5):
        games = games[games['bot_{}'.format(i)].isin(bots)]
    
    games.reset_index(drop=True, inplace=True)
    
    new = []
    for same_or_different in itertools.product([True, False], repeat=5):
        new_df = pd.DataFrame()
        new_df['winner'] = games.winner
        new_df['win_type'] = games.win_type
        for player in range(5):
            new_df['bot_{}_role'.format(player)] = games['bot_{}_role'.format(player)]
            new_df['bot_{}_payoff'.format(player)] = games['bot_{}_payoff'.format(player)]
        
        for player, same in enumerate(same_or_different):
            if same:
                bot_name = games['bot_{}'.format(player)].astype(str)
                new_df['bot_{}'.format(player)] = (
                    bot_name + ":" + bot_name
                ).astype('category')
            else:
                all_series = []
                for bot, other in [bots, bots[::-1]]:
                    filtered = games[games['bot_{}'.format(player)] == bot]
                    bot_role = filtered['bot_{}_role'.format(player)]
                    bot_name = filtered['bot_{}'.format(player)].astype(str)
                    
                    is_resistance_series = bot_name + ":{}".format(other)
                    is_spy_series = "{}:".format(other) + bot_name
                    new_series = np.where(bot_role.isin(['merlin', 'servant']), is_resistance_series, is_spy_series)
                    new_series = pd.Series(new_series, index=bot_role.index)
                    all_series.append(new_series)
                new_df['bot_{}'.format(player)] = pd.concat(all_series).sort_index().astype('category')
        
        new.append(new_df)
        
    result = pd.concat(new)
    result.reset_index(drop=True, inplace=True)
    return result

from scipy.special import binom
def get_counts(all_counts, res_counts):
    return (
        binom(all_counts[0], res_counts[0]) *
        binom(all_counts[1], res_counts[1]) *
        binom(all_counts[2], res_counts[2]) *
        binom(all_counts[3], res_counts[3])
    )

def get_prob(subcounts, index, total_counts, expected_counts):
    all_counts = index[:4]
    res_counts = index[4:]
    
    possible_ways_to_sort = expected_counts[all_counts]
    possible_ways_to_sort_this_way = get_counts(all_counts, res_counts)
    expected_proportion = float(possible_ways_to_sort_this_way)/possible_ways_to_sort
    
    num_games = subcounts
    total_counts = total_counts[all_counts]
    actual_proportion = float(num_games)/total_counts
    
    return expected_proportion/actual_proportion
    

from collections import defaultdict
def splat_by_role(games, bots):
    assert len(bots) == 2
    print "Splatting"
    generated_games = splat_generate_games(games, bots)
    expanded_bots = generated_games.bot_0.unique()
    print "Computing counts"
    for bot in expanded_bots:
        generated_games['{}_res'.format(bot)] = 0
        generated_games['{}_res'.format(bot)] = generated_games['{}_res'.format(bot)].astype(int)
        generated_games['{}_all'.format(bot)] = 0
        generated_games['{}_all'.format(bot)] = generated_games['{}_all'.format(bot)].astype(int)
        
    for i in range(5):
        is_res = generated_games['bot_{}_role'.format(i)].isin(['servant', 'merlin'])
        for bot in expanded_bots:
            is_bot = generated_games['bot_{}'.format(i)] == bot
            generated_games['{}_res'.format(bot)] += (is_res & is_bot).astype(int)
            generated_games['{}_all'.format(bot)] += (is_bot).astype(int)
    
    print "Indexing"
    index_names = ['{}_all'.format(bot) for bot in expanded_bots] + ['{}_res'.format(bot) for bot in expanded_bots]
    generated_games.set_index(index_names, inplace=True)
    
    counts = defaultdict(int)
    possible_counts = defaultdict(float)
    print generated_games.index.values[0]
    possible_index_values = list(set(generated_games.index.values))
    for index_value in possible_index_values:
        all_counts = index_value[:4]
        possible_counts[all_counts] += get_counts(all_counts, index_value[4:])
        counts[all_counts] += len(generated_games.loc[index_value])
    
    for index_value in possible_index_values:
        prob = get_prob(len(generated_games.loc[index_value]), index_value, counts, possible_counts)
        generated_games.loc[index_value, 'keep_percent'] = prob
    
    generated_games.reset_index(inplace=True)
    generated_games.set_index(['{}_all'.format(bot) for bot in expanded_bots], inplace=True)
    
    for index_value in list(set(generated_games.index.values)):
        generated_games.loc[index_value].keep_percent /= generated_games.loc[index_value].keep_percent.max()
        
    print "Filtering"
    generated_games.reset_index(inplace=True)
    keep = generated_games.keep_percent > np.random.random(len(generated_games))
    generated_games = generated_games[keep]
    generated_games.reset_index(drop=True, inplace=True)
    return generated_games
```

```python
def sel(name, role, res, spy):
    return (
        (name.isin(res) & role.isin(['merlin', 'servant'])) |
        (name.isin(spy) & ~(role.isin(['merlin', 'servant'])))
    )
```

```python
df = ALL_GAMES
res_bots = ['Deeprole', 'ObserveBot']
spy_bots = res_bots

select = (
    sel(df.bot_0, df.bot_0_role, res_bots, spy_bots) &
    sel(df.bot_1, df.bot_1_role, res_bots, spy_bots) &
    sel(df.bot_2, df.bot_2_role, res_bots, spy_bots) &
    sel(df.bot_3, df.bot_3_role, res_bots, spy_bots) &
    sel(df.bot_4, df.bot_4_role, res_bots, spy_bots)
)

random_res_games = df[select]
random_res_games.set_index(['bot_0'], inplace=True)
random_res_games.index.values.unique()
```
